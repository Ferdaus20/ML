```python
x = df['x']
y = df['y']
```

linear reg using gradient decent

```python
m = 0
c = 0
lr = .0001
n = len(df.x)


for i in range(500):
    y_hat= (m*x)+c
    dm = (-2/n)*sum(x*(y-y_hat))
    dc = (-2/n)*sum(y-y_hat)


    m -= lr*dm
    c -= lr*dc
    print(m,c)

print(m,c)
    
```
